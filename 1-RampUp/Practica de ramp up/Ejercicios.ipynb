```python
def traducir_pais(codigo):
    dicc = {
    "es": "Espanha",
    "us": "Estados Unidos",
    "fr": "Francia"}
    codigo = codigo.lower()
    return dicc.get(codigo,"no esta act")
print(traducir_pais("Vz"))
```

```python


def que_hacer_hoy():
    dia=input("que dia es hoy?")
    rutina={
        'lunes' :  'piernas',
        'martes' : 'brazos',
        'miercoles' : 'abdomen',
        'jueves' : 'cardio',
        'viernes' : 'cerveza',
        'sabado' : 'pizza',
        'domingo' : 'descansa',}
    try:
        return print(rutina[dia])
    except KeyError:
        return print('no es correcto')
que_hacer_hoy()

```

```python
def contar_vocales(c:str):
    contador=0 
    for letra in c.lower():
        if letra in "aeiou" :
            contador+=1
    return contador
```

```python
import string
import random
def  generar_contrasenha(longitud:int) -> str:
    caracteres = string.ascii_letters + string.digits + string.punctuation
    contrasenha = "".join(random.choice(caracteres) for _ in range(random.randint(8,16)))
    return contrasenha
print(generar_contrasenha)
```

```python
numeros=["a","2","3","4","5","6","7","8","9","10""j","q","k"] # numero de cartas
palos = ["♥","♦","♣","♠"] # los simbolos se colocan precionando alt = 3,4,5,6

baraja = []

for numero in numeros: # bucle for
    for palo in palos:
        baraja.append(f'{numero}{palo}')
random.sample(baraja,5)
```
